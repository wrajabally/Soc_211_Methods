### 3. Create a bar graph with confidence interval whiskers

Create a bar graph that compares interval ratio data for a dependent variable across different nominal categories from an independent variable. If you are using Charlie's data, do this for the state grant aid per low income student variable name **sgrnt_low**

```stata
rename (v27 v28) (tov_prcnt orb_prcnt)
```

```stata
sum efg tov_prcnt orb_prcnt ftfga
gen tovrate = tov_prcnt/100
gen orbprcnt = orb_prcnt/100
gen fourfactor = (.4 * efg) + (.25 * tovrate) + (.2 * orbprcnt) + (.15 * ftfga)
tab fourfactor
```

```stata
gen fourfactorRnk = 1 if fourfactor<=.3122
replace fourfactorRnk = 2 if fourfactor>.3122 & fourfactor<=.32
replace fourfactorRnk = 3 if fourfactor>.32
label define fourfactorRnkl 1 "Bottom Third" 2 "Middle Third" 3 "Upper Third" 
label value fourfactorRnk fourfactorRnkl

```

```stata
tab fourfactor fourfactorRnk
```

```stata
twoway (bar w fourfactorRnk) (rcap w sos fourfactorRnk), ///
title("Wins compared with the Four Factor Measure according to SOS, '18-'19", size(medium)) ///
xlabel(1 "Bottom Third" 2 "Middle Third" 3 "Top Third") ///
ytitle("Number of Wins 2018-2019") ///
xtitle("Rank of Four Factor Measure")

/* graph twoway (bar mn_igrnt_low hgroup) ///
(rcap hiigrnt_low loigrnt_low hgroup), ///
xlabel( 1 "Highest" 3 "Very High" 5 "High" 7 "Some/None", noticks) ///
ytitle("Grant aid per Pell recipient (2015 $)") ///
ylabel(0(5000)20000, format(%10.0fc)) ///
xtitle("Research Category") legend(off) name(igrntbar, replace) ///
title("Institutional Grant Aid") ///
yline(1936.825, lp(_) lcolor(red) lstyle(foreground)) */
```
